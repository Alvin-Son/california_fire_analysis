```python
import pandas as pd
```

```python
# Create csv read range lists
years = [year for year in range(2020, 2025)]
csv_numbers = [month for month in range(1, 13)]

# Create empty DataFrame for combined data
rainfall_all = pd.DataFrame()

# Loop through all years and months of data, month and year form header and ad as new column for each set
for year in years:
    for num in csv_numbers:
        
        try:
            base_path = f"Resources/weather_data/rainfall/{year}/data ({num}).csv" 

            header_row = pd.read_csv(base_path, nrows=0).columns[0]
            rainfall_data = pd.read_csv(base_path, skiprows=4)

            county_date = header_row.split()
            rainfall_month = county_date[0]
            rainfall_year = county_date[1]

            rainfall_data["Month"] = rainfall_month
            rainfall_data["Year"] = rainfall_year
            rainfall_all = pd.concat([rainfall_all, rainfall_data], ignore_index=True)
        
        except FileNotFoundError as e:
            print(e)
        
rainfall_all
```

```python
# Check to see if all data has enteres correctly
rainfall_all.info()
```

```python
# Insert column for County
rainfall_all["County"] = [row.split()[0] for row in rainfall_all["Name"]]
```

```python
# Remove unnecessary columns
del rainfall_all["Rank"]
del rainfall_all["Name"]
rainfall_all
```
