```python
import pandas as pd
```

```python
# Create csv read range lists
years = [year for year in range(2013, 2025)]
csv_numbers = [month for month in range(1, 13)]

# Create empty DataFrame for combined data
temperature_all = pd.DataFrame()

# Loop through all years and months of data, month and year form header and ad as new column for each set
for year in years:
    for num in csv_numbers:
        
        try:
            base_path = f"Resources/weather_data/max_temp/{year}/data ({num}).csv" 

            header_row = pd.read_csv(base_path, nrows=0).columns[0]
            temperature_data = pd.read_csv(base_path, skiprows=4)

            county_date = header_row.split()
            temperature_month = county_date[0]
            temperature_year = county_date[1]

            temperature_data["Month"] = temperature_month
            temperature_data["Year"] = temperature_year
            temperature_all = pd.concat([temperature_all, temperature_data], ignore_index=True)
        except FileNotFoundError as e:
            print(e)
        
temperature_all
```

```python
# Check to see if all data has enteres correctly
temperature_all.info()
```

```python
# Insert column for County
temperature_all["County"] = [row.rsplit(' ', 1)[0] for row in temperature_all["Name"]]
temperature_all
```

```python
# Select a data subset for rainfall
temperature_clean = temperature_all[['County','Value','Month','Year']].rename(columns={'Value':'Temperature (F)'})
temperature_clean.to_csv("Resources/weather_data/temperature_all_years_clean.csv", index_label="City_ID")
```

```python
temp_duplicates = temperature_clean[["County","Month", "Year"]].duplicated()
temperature_clean[temp_duplicates]
```

```python
county_year_ave_temperature_df = temperature_clean.groupby(["County", "Year"])["Temperature (F)"].mean().to_frame().reset_index()
county_year_ave_temperature_df
```

```python
county_month_ave_temperature_df = temperature_clean.groupby(["County", "Month"])["Temperature (F)"].mean().to_frame().reset_index()
county_month_ave_temperature_df
```

```python
county_10yr_ave_temperature_df = temperature_clean.groupby(["County"])["Temperature (F)"].mean().to_frame().reset_index()
county_10yr_ave_temperature_df.sort_values("Temperature (F)", ascending=False)
```
