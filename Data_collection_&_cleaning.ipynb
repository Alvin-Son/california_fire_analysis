## Total Monthly Rainfall Data

```python
import pandas as pd
```

```python
# Create csv read range lists
years = [year for year in range(2013, 2025)]
csv_numbers = [month for month in range(1, 13)]

# Create empty DataFrame for combined data
rainfall_all = pd.DataFrame()

# Loop through all years and months of data, month and year form header and ad as new column for each set
for year in years:
    for num in csv_numbers:
        
        try:
            base_path = f"Resources/weather_data/rainfall/{year}/data ({num}).csv" 

            header_row = pd.read_csv(base_path, nrows=0).columns[0]
            rainfall_data = pd.read_csv(base_path, skiprows=4)

            county_date = header_row.split()
            rainfall_month = county_date[0]
            rainfall_year = county_date[1]

            rainfall_data["Month"] = rainfall_month
            rainfall_data["Year"] = rainfall_year
            rainfall_all = pd.concat([rainfall_all, rainfall_data], ignore_index=True)
        
        except FileNotFoundError as e:
            print(e)
        
rainfall_all
```

```python
# Check to see if all data has enteres correctly
rainfall_all.info()
```

```python
# Insert column for County
rainfall_all["County"] = [row.rsplit(' ', 1)[0] for row in rainfall_all["Name"]]
rainfall_all
```

```python
# Select a data subset for rainfall
rainfall_clean = rainfall_all[['County','Value','Month','Year']].rename(columns={'Value':'Rainfall (inches)'})
rainfall_clean.to_csv("Resources/weather_data/rainfall_all_years_clean.csv", index_label="City_ID")

# Total check: 58 counties * 11.25 years * 12 months = 7830 rows
# Show final rainfall data
rainfall_clean
```

## Fire Incident Data

```python
import pandas as pd
from pathlib import Path
import calendar
import warnings

# Ignore warnings
warnings.filterwarnings("ignore")

# Import raw fire data
file_path=Path("Resources/fire_data/mapdataall.csv")
df=pd.read_csv(file_path, encoding="utf-8")

df.head()
```

```python
# Create list of original columns for reference
list(df.columns)
```

```python
# Create new data frame with select columns
fire_raw = df[['incident_name',
 'incident_administrative_unit',
 'incident_county',
 'incident_acres_burned',
 'incident_longitude',
 'incident_latitude',
 'incident_dateonly_extinguished',
 'incident_dateonly_created'
]]

# Rename columns
fire_raw = fire_raw.rename(columns={
    'incident_name': "Name",
     'incident_administrative_unit': "Admin unit",
     'incident_county': "County",
     'incident_acres_burned': "Acres burned" ,
     'incident_longitude': "Lon",
     'incident_latitude': "Lat",
     'incident_dateonly_extinguished': "Date extinguished",
     'incident_dateonly_created': "Date created"
})

# Show raw fire dataframe
fire_raw.head()
```

```python
# Drop rows where county and acres burned columns are null
fire_clean = fire_raw.dropna(subset = ['Acres burned','County', 'Date extinguished'])

# Convert to datetime
fire_clean['Date extinguished'] = pd.to_datetime(fire_clean.loc[:,'Date extinguished'])
fire_clean['Date created'] = pd.to_datetime(fire_clean.loc[:,'Date created'])

# Calculate duration
fire_clean['Duration (days)'] = fire_clean.loc[:,'Date extinguished'] - fire_clean.loc[:,'Date created']

# Add years columns. Note dates are currently in datetime
fire_clean["Year extinguished"] = fire_clean.loc[:,"Date extinguished"].dt.year
fire_clean["Year"] = fire_clean.loc[:,"Date created"].dt.year

# Add months column by name number
fire_clean["Month created (num)"] = fire_clean["Date created"].dt.month
fire_clean["Month"] = fire_clean["Month created (num)"].apply(lambda x: calendar.month_name[x])

# Show clean dataframe
fire_clean.head()
```

```python
# Check for nulls
fire_clean.info()
```

```python
# Select desired analysis columns
fires_df = fire_clean[["Name","Year", "Month","Month created (num)", "County", "Acres burned", "Duration (days)", "Lat", "Lon"]]

# Drop "days" from duration and convert to float
fires_df["Duration (days)"] = fires_df["Duration (days)"].astype(str)
fires_df.loc[:, "Duration (days)"] = pd.to_numeric(fires_df["Duration (days)"].str.extract('(\d+)')[0], errors="coerce")
fires_df["Duration (days)"] = fires_df["Duration (days)"].astype(float)

# Show selected fire columns dataframe
fires_df.head()
```

```python
# Collect new dataframe slice of rows with multiple counties for single fire
fire_duplicate_df = fires_df.loc[fires_df['County'].str.contains(','), :]
fire_duplicate_df.head()
```

```python
# Create function to split multiple counties and divide acres burned equally between them
def expand_counties(row):
    counties = row['County'].split(", ")
    num_counties = len(counties)
    # Create new row for each county with the duration unchanged
    return [
        {**row, 'County': county, 'Acres burned': row['Acres burned'] / num_counties, 'Duration (days)': row['Duration (days)']}
        for county in counties
    ]

# Apply the function to each row and expand into a list of dictionaries
expanded_rows = [new_row for index, row in fire_duplicate_df.iterrows() for new_row in expand_counties(row)]

# Convert the list of dictionaries to a DataFrame
split_county_df = pd.DataFrame(expanded_rows)

# Show new split county dataframe
split_county_df
```

```python
# Create a set of multiple county dataframe slice "County" column
counties_to_drop = set(fire_duplicate_df['County'])

# Create list of unique rows to drop using counties_to_drop set from original dataframe
mask = ~fires_df['County'].isin(counties_to_drop)

# Save final dataframe with unwanted multiple county ropws filtered out
fire_data = fires_df[mask]

# Confirm original number of rows 1943 - 39 duplicate rows = 1904

# Create final clean raw dataframe by merging new split counties and filtered counties dataframes
fires_all_ungrouped = pd.concat([fire_data, split_county_df], axis=0, ignore_index=True)
fires_all_ungrouped

# Export clean fires grouped dataframe for mapping before removing Lat and Long
fires_all_ungrouped.to_csv("Resources/fire_data/fire_all_ungrouped.csv", index_label="City_ID")

# Check that the row count is accurate: 99 + 1904 = 2003
fires_all_ungrouped
```

```python
# Create grouped dataframe by Year, Month and County
fire_grouped = fires_all_ungrouped.groupby(['Year','Month','Month created (num)', 'County'])

## Apply aggregations
# Calculate average acres burned
acres_burned_avg = fire_grouped["Acres burned"].mean()

# Calculate total acres burned
acres_burned_tot = fire_grouped["Acres burned"].sum()

# Calculate average duration in days
duration_days_avg = fire_grouped["Duration (days)"].mean()

#Calculate sum of total incidents
tot_fire_incidents = fire_grouped["Acres burned"].count()

# Save to new dataframe
summary_fire = pd.DataFrame({
    "Acres Burned Avg": acres_burned_avg,
    "Acres Burned Tot": acres_burned_tot,
    "Fire Duration Avg (days)": duration_days_avg,
    "Incident Count Tot": tot_fire_incidents
 }).reset_index()

# Filter out unwanted years
fire_all_grouped = summary_fire.loc[(summary_fire["Year"]>=2013) & (summary_fire["Year"]<2024)]

# Export to csv for merging with weather data
fire_all_grouped.to_csv("Resources/fire_data/fire_all_grouped.csv", index_label="City_ID")
fire_all_grouped
```
